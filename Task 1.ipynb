```python
#filled null in age with median of age column
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')  # ensure numeric
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
#filled null in date with random date
df['JoinDate'] = pd.to_datetime(df['JoinDate'], errors='coerce')
df['JoinDate'] = df['JoinDate'].fillna(pd.to_datetime("2000-01-01"))
df
```

```python
# Ensuring the column is numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
# Filled nulls with the minimum salary
df['Salary'] = df['Salary'].fillna(df['Salary'].min())
df
```

```python
#changed datatypes
df['Age'] = df['Age'].astype(int)
df['Salary'] = df['Salary'].astype(int)
df.dtypes
```

```python
#dropped the email which had null
df = df.dropna(subset=['Email'])
df
```

```python
import numpy as np
df['Gender'] = df['Gender'].replace({
    'M': 'Male',
    'F': 'Female',
    'Unknown': 'Female'
})
# Replace actual NaN (not string) with a value (e.g., 'Male')
df['Gender'] = df['Gender'].fillna('Male')
df
```
